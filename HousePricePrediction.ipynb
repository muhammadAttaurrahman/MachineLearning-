## Filling Null Values

```python
# There are three ways to fill N/A values 
# 1 ) Remove the missing values 
# 2) remove the entire attribute 
# 3) add mean to the missing values 

a = housing_data.dropna(subset=['RM']).shape #option1
housing_data.drop("RM", axis=1).shape #option2
#Note that the original dataframe wont be changed
median = housing_data["RM"].median()
housing_data["RM"].fillna(median) #option3
```

```python
housing_data = strat_train_set.drop("MEDV", axis=1)
housing_data.shape
```

```python
housing_lables = strat_train_set['MEDV'].copy()
```

## Introduction to Imputer class

```python
#now if there are multiple values missing in the data frame
#so we have yo use imputer class to fill them
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
imputer.fit(housing_data)
```

## Creating Pipeline 

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline ( [
    ('imputer', SimpleImputer (strategy ='median')),
                         #add as many you want arguments
    ('std_scaler', StandardScaler()),
]) 
```

```python
housing_data_num = my_pipeline.fit_transform(housing_data)
```

```python
housing_data_num
#it becomes a numpy aray 
```

## Model Selection 

```python
#as it is nomaric data so we will use Linear Regression 
# you can change the model as wel 
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
model = DecisionTreeRegressor()
model = RandomForestRegressor()
# model = LinearRegression()
model.fit(housing_data_num, housing_lables)
```

```python
# Now if you want to chyeck your model fitting, try on some dta 
some_data =  housing_data.iloc[0:5]
some_lables= housing_lables.iloc[:5]
prepared_data = my_pipeline.transform(some_data)
model.predict(prepared_data)
```

```python
# list(some_lables)
housing_data_num
```

## Evaluating the Model 

```python
# We will find mean squreed error which is used for the avergae error between actuall values and 
# predicted values, the less it is, the best our model would be
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(housing_data_num)
mse = mean_squared_error(housing_lables,housing_predictions)
rmse = np.sqrt(mse)
```

```python
rmse
```

## Using better evaluation Techniques, cross validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model , housing_data_num , housing_lables, scoring="neg_mean_squared_error")
rmse_score = np.sqrt(-scores)
```

```python
rmse_score
```

```python
def prnt_scores (scores):
    print ("SCores", scores)
    print ("Mean", scores.mean())
    print ("Std", scores.std())
```

```python
prnt_scores(rmse_score)
```
